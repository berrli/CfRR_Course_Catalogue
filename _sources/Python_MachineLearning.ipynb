## Introduction to Machine Learning

Machine Learning (ML) is a subfield of Artifical Intelligence (AI), a branch of computer science that looks to create system capable of performing tasks that normally require typically human intelligence {cite}`mccarthy:2006:AIDefinition`. ML aims to enable computer to perform a range of tasks without explicit instruction, formally defined as "A computer program is said to learn from experience E concerning some class of tasks T and performance measure P if its performance at tasks in T, as measured by P, improved with experience E." {cite}`Mitchell:1997:MLDefinition`.
